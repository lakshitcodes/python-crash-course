```python
import pandas as pd
import numpy as np
# import csv file
my_df = pd.read_csv('sample_data.csv')
my_df
```

```python
# Count distinct values - Descending Order
my_df['Name'].value_counts()
```

```python
# Count distinct values - Ascending Order
my_df['HEX'].value_counts(ascending=True)
```

```python
# Names without NaN
my_df['Name'].value_counts()
```

```python
# Names with NaN
my_df['Name'].value_counts(dropna=False)
```

```python
# Get relative frequency - percentage
my_df['HEX'].value_counts(normalize=True)
```

```python
# Get specific item count
my_df['Name'].value_counts()['White']
```

```python
# Count Unique values - size
my_df.groupby('Name').size()
```

```python
# Count unique values - count
my_df.groupby('Name').count()
```

```python
# Get a count all columns across all columns
my_df.apply(pd.value_counts)
```
