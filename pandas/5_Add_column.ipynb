```python
import pandas as pd
import numpy as np
# Import CSV file
my_df = pd.read_csv("sample_data.csv")
my_df
```

```python
# Add column from list
index = np.arange(17)
my_df["index"]=index
my_df
```

```python
# Add default values
my_df['Available']=[True]*len(my_df)
my_df
```

```python
# Add a NaN Value
my_df["opacity"]=[np.nan]*len(my_df)
my_df
```

```python
# Add column with .insert() - allows position
my_df.insert(2,"Natural",[True]*len(my_df))
my_df
```

```python
# Add column with .assign() - creates new dataframe
my_df2 = my_df.assign(Synthetic=[False]*len(my_df))
my_df2
```
