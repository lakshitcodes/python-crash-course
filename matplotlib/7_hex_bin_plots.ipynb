```python
import numpy as np
import pandas as pd
from numpy.random import randn
%matplotlib inline

# Create dataframe
my_df = pd.DataFrame(randn(500,4),columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
# Hexplots
my_df.plot(kind='hexbin',x='Mon',y='Tues')
```

```python
# Hexplots grid size
# Bigger grid size -> Smaller hexagon size
my_df.plot(kind='hexbin',x='Mon',y='Tues',gridsize=25)
```

```python
# Hexplots color
my_df.plot(kind='hexbin',x='Mon',y='Tues',gridsize=10,cmap='magma')
```

```python
# Other method
my_df.plot.hexbin(x='Mon',y='Tues',gridsize=10)
```

```python
# Title
my_df.plot.hexbin(x='Mon',y='Tues',gridsize=10,title="my hexbin plot")
```

```python
# Grid
my_df.plot.hexbin(x='Mon',y='Tues',gridsize=10,grid=True)
```
