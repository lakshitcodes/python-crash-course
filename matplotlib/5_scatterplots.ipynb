```python
import numpy as np
import pandas as pd
from numpy.random import randn
%matplotlib inline

# Create dataframe
my_df = pd.DataFrame(randn(500,4),columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
# Scatter plot
my_df.plot(kind='scatter',x='Mon',y='Tues',c='Wed')
```

```python
# Colors : https://matplotlib.org/stable/users/explain/colors/colormaps.html
my_df.plot(kind='scatter',x='Mon',y='Tues',c='Wed',cmap='magma')
```

```python
# By size
my_df.plot(kind='scatter',x='Mon',y='Tues',s=my_df['Wed']*100)
```

```python
# By size easier to read with alpha
my_df.plot(kind='scatter',x='Mon',y='Tues',s=my_df['Wed']*100,alpha=0.4)
```

```python
# Add title
my_df.plot(kind='scatter',x='Mon',y='Tues',s=my_df['Wed']*100,alpha=0.4,title="My Awesome Scatter plot")
```
