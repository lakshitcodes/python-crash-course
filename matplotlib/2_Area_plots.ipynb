```python
import pandas as pd 
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(25,4),columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
# area plot
my_df.abs().plot(kind='area')
```

```python
# Shading
my_df.abs().plot(kind='area',alpha=0.5)
```

```python
# Title
my_df.abs().plot(kind='area',alpha=0.5,title="My Awesome Area Plot")
```

```python
# Legend
my_df.abs().plot(kind='area',alpha=0.5,title="My Awesome Area Plot",legend=False)
```

```python
# Plot only 1 column
my_df['Mon'].abs().plot(kind='area',alpha=0.5,title="My Awesome Area Plot",legend=False)
```

```python
# Plot 2 columns
my_df[['Mon','Thurs']].abs().plot(kind="area",alpha=0.5,legend=True,title="My arae plot")
```

```python
# Add Table
my_df.abs().plot(kind="area",alpha=0.5,legend=True,title="My area plot",table=True)
```
