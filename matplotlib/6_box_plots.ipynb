```python
import numpy as np
import pandas as pd
from numpy.random import randn
%matplotlib inline

# Create dataframe
my_df = pd.DataFrame(randn(500,4),columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
# Box plots
my_df.plot(kind='box')
```

```python
# boxplots size
my_df.plot(kind='box',figsize=(10,5))
```

```python
# Boxplots color
my_df.plot(kind='box',cmap='jet')
```

```python
# Legend
my_df.plot(kind='box',legend=True)
```

```python
# Title
my_df.plot(kind='box',title="My box plot")
```

```python
# Other way
my_df.plot.box()
```

```python
# Add grid
my_df.plot(kind='box',title="My box plot",grid=True)
```
