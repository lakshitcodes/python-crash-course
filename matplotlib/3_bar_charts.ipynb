```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
# Create Dataframe
my_df = pd.DataFrame(randn(25,4),columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
# bar plot
my_df.abs().plot(kind='bar')
```

```python
# bar plot stacked
my_df.abs().plot(kind='bar',stacked=True)
```

```python
# Shading
my_df.abs().plot(kind='bar',stacked=True,alpha=0.4)
```

```python
# Title, Legend, can be used like in the previous file
```

```python
# Second Way
my_df.abs().plot.bar()
```

```python
# Second Way stacked
my_df.abs().plot.bar(stacked=True)
```

```python
# Second Way shading
my_df.abs().plot.bar(stacked=True,alpha=0.5)
```

```python
# Second Way title
my_df.abs().plot.bar(stacked=True,alpha=0.5,title="My bar chart")
```

```python
# Second Way Legend
my_df.abs().plot.bar(stacked=True,alpha=0.5,title="My bar chart",legend=False)
```

```python
# Grid lines
my_df.abs().plot.bar(stacked=True,alpha=0.5,title="My bar chart",grid=True)
```
