```python
import numpy as np
import pandas as pd
from numpy.random import randn
%matplotlib inline

# Create dataframe
my_df = pd.DataFrame(randn(25,4),columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
# Line plot
my_df.plot(kind='line')
```

### Line plot absolute
my_df.abs().plot(kind='line')

```python
# Line width
my_df.abs().plot(kind='line',lw=3)
```

```python
# Line single
my_df['Mon'].abs().plot(kind='line')
```

```python
# Line more than one
my_df[['Mon','Wed']].abs().plot(kind='line')
```

```python
# Size
my_df.abs().plot(kind='line',figsize=(10,5))
```

```python
# Title
my_df['Mon'].abs().plot(kind='line',title='my line chart')
```

```python
# Shading ,legend like others
```

```python
# Second way
my_df.abs().plot.line()
```
